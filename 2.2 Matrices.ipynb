```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Two images for blending
img1 = mpimg.imread('apple.jpg')
img2 = mpimg.imread('leaf.jpg')

# Some preprocessing realted to shape
img2 = img2[:img1.shape[0], :img1.shape[1], :]

# Blend the images with a specified weight
weight = 1.5 
blended_image = weight * img1 + (0.7 - weight) * img2


```

```python
# Display the original images and the blended image
plt.figure(figsize=(10, 5))

plt.subplot(1, 4, 1)
plt.imshow(img1)
plt.title('Image 1')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(img2)
plt.title('Image 2')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(blended_image)
plt.title('Blended Image')
plt.axis('off')

plt.show()
```
