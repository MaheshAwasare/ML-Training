```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np


img_path = 'sample_face.jpg'

img = mpimg.imread(img_path)
print(img)

# Display the original image
plt.imshow(img, cmap='gray')  
plt.axis('off')
plt.title('Original Image')
plt.show()


```

```python
import matplotlib.pyplot as plt
import numpy as np 
from skimage import io
from sklearn.decomposition import PCA


img = io.imread('apple.jpg')  


# Define utility function to display images
def show_image(img, title=''):
    plt.imshow(img)
    plt.title(title)
    plt.axis('off')
    plt.show()

# PCA reconstruction
n_components = 2
image_vector = img.reshape(-1, img.shape[-1])  

pca = PCA(n_components=n_components)
img_compressed = pca.fit_transform(image_vector)  

img_approx = pca.inverse_transform(img_compressed) 
img_approx = img_approx.reshape(img.shape)

# Display original and reconstructed images   
show_image(img, 'Original Image')
show_image(img_approx, 'Reconstructed Image with PCA')

print(f"{n_components} components extracted!")
```
